
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.
<br>
import numpy as np<br>
import matplotlib.pyplot as plt<br>
import pandas as pd<br>
hollywood_actor_kills = pd.read_csv('actor_kill_counts.csv')<br>
hollywood_actor_kills=hollywood_actor_kills.sort_values(by=['Count'],ascending=False)<br>
actor_names = tuple(hollywood_actor_kills['Actor'].values.tolist())<br>
kill_counts = hollywood_actor_kills['Count'].values.tolist()<br>
hollywood_actor_kills<br>
objects = actor_names<br>
y_pos = np.arange(len(objects))<br>
performance = kill_counts<br>
plt.barh(y_pos, performance, align='center', alpha=0.5)<br>
plt.yticks(y_pos, objects)<br>
plt.xlabel('Kills')<br>
plt.title('Deadliest actors in Hollywood')<br>
plt.show()
